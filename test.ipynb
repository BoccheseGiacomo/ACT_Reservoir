```python
import numpy as np
from scipy import optimize
import matplotlib.pyplot as plt
```

```python
W=np.array([   
    [1,-1],
    [1,0]
])

x0=np.array([
    [1],
    [0]
])
```

```python
x = x0
x_series = [x0.flatten()]  # Using a list to store the series of x, and flattening x0 for easier plotting

for _ in range(10):
    x = W @ x
    x_series.append(x.flatten())  # Flatten x for easier plotting
    print(x)
    print()

# Convert the list of x series into a NumPy array for easier slicing
x_series = np.array(x_series)

# Plot the series of the first element of x
plt.plot(x_series[:, 0], label='First element of x')

# Plot the series of the second element of x
plt.plot(x_series[:, 1], label='Second element of x')

plt.xlabel('Iteration')
plt.ylabel('Value')
plt.title('Series of Each Element of Array x')
plt.legend()
plt.show()
```
